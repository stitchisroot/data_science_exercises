Scatter plots are used to display the relationship between two numerical variables in data visualization. Seaborn librart in Python is quite helpful in generating such scatter plots https://seaborn.pydata.org/generated/seaborn.scatterplot.html. Using the following mpg dataset from Seaborn's data repository, create a scatter plot of miles per gallon against against horsepower. In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. Also, transparency of a plot is managed by the parameter alpha between 0 an 1. Use these parameters in your scatter plot to enrich your display.

```python
import seaborn as sns
sns.set(style="white")
```

```python
# Answer:
mpg = sns.load_dataset("mpg")
sns.relplot(x="horsepower", y="mpg", hue="origin", size="weight",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=mpg)
```

## Q3.
Once you have a cleaned dataset, the next step is Exploratory Data Analysis (EDA). It is a process of exploring patterns, relationship, or anomalies to guide us in our next steps in our data analysis. One of the important tools that we utilize in EDA is the pair plots - a scatterplot matrix. Pair plots display us the distribution of a single variable along with pairwise relationships between variables in our data set. Using the paiplot function in Seaborn library (https://seaborn.pydata.org/generated/seaborn.pairplot.html), explore the following socioeconomic data set obtained from Gapminder (https://www.gapminder.org/data/). Utilize the hue parameter in your pairplots. What relationships do you identify between pairs of columns?


```python
import pandas as pd
url = 'https://media.githubusercontent.com/media/WillKoehrsen/Data-Analysis/master/pairplots/data/gapminder_data.csv'
data = pd.read_csv(url)
print(data)
```

```python
# Import Seaborn visualization library
import seaborn as sns
sns.set(style="white")
```

```python
# Answer:
sns.pairplot(data, hue = 'continent', diag_kind = 'kde',
             plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'},
             size = 4)
```
